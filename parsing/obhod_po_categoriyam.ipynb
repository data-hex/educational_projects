```python
from bs4 import BeautifulSoup
import requests
```

```python
class Parser:
    def __init__(self):
        self.prices = []
        self.categories = []
        self.counter = 0
    
    def get_response(self, url):
        response = requests.get(url=url)
        response.encoding = 'utf-8'
        return response
    
    def page_visit(self, vpage):
        response = self.get_response(vpage)
        soup = BeautifulSoup(response.text, 'lxml')
        print("Посещаю страницу товара")
        price = soup.find('span', id='price').text
        price = price.split()
        count = soup.find('span', id='in_stock').text
        count = count.split()
        return int(price[0])*int(count[2])       
        
    
    def list_pages(self, response, site):
        soup = BeautifulSoup(response.text, 'lxml')
        pagen = [f"{site}{link['href']}" for link in soup.find('div', class_='pagen').find_all('a')]
        print("Страницы пагинации найдены")
        return pagen
    
    def list_categories(self, response,site):
        soup = BeautifulSoup(response.text, 'lxml')
        categories = [f"{site}{link['href']}" for link in soup.find('div', class_='nav_menu').find_all('a')]
        return categories
    
    def start_parse(self, url, site):
        response = self.get_response(url)
        #self.categories = self.list_categories(response, site)
        for page in self.list_pages(response, site):
            soup = BeautifulSoup(self.get_response(page).text, 'lxml')
            vpages = [f"{site}{link['href']}" for link in soup.find_all('a', class_='name_item')]
            for vpage in vpages:
                self.prices += [int(self.page_visit(vpage))]
        
        
    def visit_categories(self,url,site):
        response = self.get_response(url)
        self.categories = self.list_categories(response, site)
        for i in self.categories:
            url = i
            print('Захожу в категорию')
            self.start_parse(url,site)         
        
        print(sum(self.prices))
        

```

```python
if __name__ == '__main__':
    url = 'https://parsinger.ru/html/index3_page_4.html'
    site = 'http://parsinger.ru/html/'
    parser = Parser()
    parser.visit_categories(url, site)
```
