```python
import time
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.webdriver.common.by import By
 
list_input = []
 
total = 0
 
with webdriver.Chrome() as browser:
    browser.get("https://parsinger.ru/infiniti_scroll_1")
    time.sleep(0.5)
    
    while True:
        input_text = browser.find_element(By.ID, "scroll-container").find_elements(By.TAG_NAME, "span")
        input_tags = browser.find_element(By.ID, "scroll-container").find_elements(By.TAG_NAME, "input")
    
        for item, t in zip(input_tags, input_text) :
            if item not in list_input:
                time.sleep(0.5)
                item.send_keys(Keys.TAB)
                total += int(t.text)
                print(t.text)
                item.click()
                time.sleep(0.5)
                list_input.append(item)
        time.sleep(0.5)
        
        if len(list_input) == 100:
            break
                                     
print(total)                                     
        
```
