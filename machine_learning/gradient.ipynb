```python
#task https://stepik.org/lesson/570089/step/11?unit=564632

# modules
import numpy as np
from sympy import symbols, diff, solve, Symbol, lambdify
import sympy
import matplotlib
import matplotlib.pyplot as plt

x, y = symbols('x y')
f = x*(x**2 - 9)


#starts data
h = 0.1
a0 = -1

#derivative
diff_y = f.diff(x)
diff_y = lambdify(x, diff_y)

#number of iteration steps
step = 2 

#list of coordinates
a = [a0]
count = 1

while count <= step:
    print(f'a{count} = {round(a[count-1] - diff_y(a[count-1])*h,3)}')
    a.append(round(a[count-1] - diff_y(a[count-1])*h,3))
    count +=1
print(a)

```

```python
# generation dots
xx = np.linspace(-3, 3, 200)
yy = xx*(xx**2 - 9)

# build graph
plt.figure(figsize=(10,4)) 
plt.title("Функция x*(x**2 - 9)")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xx, yy)

j = 0

for i in a:
    plt.scatter(i, round(i*(i**2 - 9), 2), color='orange', s=70, marker='o')
    plt.annotate(f'a{j}', xy=(i,round(i*(i**2 - 9), 2)),xytext=(i, round(i*(i**2 - 9), 2)+2))
    j += 1

plt.grid()
```
