```python
# modules
import numpy as np
from sympy import symbols, diff, solve
import matplotlib
import matplotlib.pyplot as plt


# generation dots
xx = np.arange(-5, 5, 0.1)
yy = xx**2

# build graph
plt.figure()
plt.title("Функция х**2")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(xx, yy)
plt.show()
```

```python
# generation dots
x = np.arange(-5, 5, 0.1)
y = x**4

# build graph
plt.figure()
plt.title("Функция х**4")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)
plt.show()
```

```python
f1 = x**2 
f2 = x**4 
x = symbols('x')
f1_derivative = diff(f1, x) 
f1_derivative
```

```python
f2_derivative = diff(f2, x) 
f2_derivative
```

```python
# generation dots
x = np.arange(-1, 1, 0.1)
y1 = 2*x
y2 = 4*x**3 

# build graph
plt.figure()
plt.title("Функции производных")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y2)
plt.plot(x, y1, '--')
plt.show()
```

```python
x = symbols('x')
inequality = 2*x - 4*x**3 > 0
solution = solve(inequality, x)
solution
```
